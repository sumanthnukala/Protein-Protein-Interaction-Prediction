```python
import requests
from bs4 import BeautifulSoup
URL = ['https://hagrid.dbmi.pitt.edu/schizo-pi/index.php/search/adv?a-all=&b-all=Schizophrenia+&a-any=&b-any=Schizophrenia&a-none=&b-none=',
'https://hagrid.dbmi.pitt.edu/schizo-pi/index.php/search/adv?p=2&a-all=&b-all=Schizophrenia&a-any=&b-any=Schizophrenia&a-none=&b-none=&sort_by=relevance',
'https://hagrid.dbmi.pitt.edu/schizo-pi/index.php/search/adv?p=3&a-all=&b-all=Schizophrenia&a-any=&b-any=Schizophrenia&a-none=&b-none=&sort_by=relevance',
'https://hagrid.dbmi.pitt.edu/schizo-pi/index.php/search/adv?p=4&a-all=&b-all=Schizophrenia&a-any=&b-any=Schizophrenia&a-none=&b-none=&sort_by=relevance',
'https://hagrid.dbmi.pitt.edu/schizo-pi/index.php/search/adv?p=5&a-all=&b-all=Schizophrenia&a-any=&b-any=Schizophrenia&a-none=&b-none=&sort_by=relevance',
'https://hagrid.dbmi.pitt.edu/schizo-pi/index.php/search/adv?p=6&a-all=&b-all=Schizophrenia&a-any=&b-any=Schizophrenia&a-none=&b-none=&sort_by=relevance']

h1 = []
for i in URL:
    page = requests.get(i)
    soup = BeautifulSoup(page.content, 'html.parser')

    for element in soup.select('td'):
        if 'and' in element.text and len((element.text).split()) == 3 and 'or' not in element.text:
            h1.append(element.text)

print(h1)
print(len(h1))
```

```python
all_seq = ''
all_seq = ' '.join(h1)
print(all_seq)
```

```python
protein = []
for i in all_seq.split():
    if i != 'and':
        protein.append(i)
#print(p1)
print(len(protein))
print(len(set(protein)))
p1 = []
p2 = []
for i in range(0,len(protein)):
    if i%2 == 0:
        p1.append(protein[i])
    else:
        p2.append(protein[i])
print(len(p1))
print(len(p2))
print(len(set(p1)))
print(len(set(p2)))
```

```python
for i in set(protein):
    print(i+'_HUMAN')
```
