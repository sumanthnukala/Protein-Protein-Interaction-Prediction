```python
#preparing the testing dataset
import pandas as pd
import requests as r
from Bio import SeqIO
from io import StringIO
import time
from Bio.SeqUtils.ProtParam import ProteinAnalysis


Pid_file = '119_SCZ_interactions.xlsx'
dfs = pd.read_excel(Pid_file)
Prt_uniprot_id = list(dfs['Uniprot_ID'])

Prt_uniprot_seq = []
for cID in Prt_uniprot_id:
    baseUrl="http://www.uniprot.org/uniprot/"
    currentUrl=baseUrl+cID+".fasta"
    response = r.post(currentUrl)
    time.sleep(2)
    cData=''.join(response.text)
    Seq=StringIO(cData)
    pSeq=SeqIO.parse(Seq,'fasta')
    for i in pSeq:
        seqid = i.id
        seq1 = str(i.seq)
        X= ProteinAnalysis(seq1) 
        W =X.molecular_weight()
        #print("%s : %0.2f Da" %(seqid,W))
        A=X.aromaticity() 
        #print("%s : %0.2f" %(seqid,A)) 
        I=X.instability_index()
        #print("%s : %0.2f" %(seqid,I))
        flex=X.flexibility()
        #print("%s :%0.2f"%(seqid,sum(flex)))
        G =X. gravy()
        #print("%s:%0.2f" %( seqid, G))
        PI=X.isoelectric_point()
        #print("%s: %0.2f " %(seqid, PI))
        print("%s: %0.2f: %0.2f: %0.2f: %0.2f: %0.2f: %0.2f " %(seqid, W,A,I,sum(flex),G,PI))
        #print('*******************')
```
